# Simple Plot Example

This is a plot of $sinc$ function

$$
f(x) = \frac{\sin(x)}{x}
$$

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
pi = np.pi
x = np.linspace(-4*pi,4*pi,1000)
plt.plot(x, np.sin(x)/x)
plt.show

## Increaseing the size and changing x and y values to LaTeX

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import numpy as np
import matplotlib.pyplot as plt
pi = np.pi
x = np.linspace(-4*pi,4*pi,1000)
plt.rcParams['figure.figsize'] = (11,4)
plt.rcParams.update({'font.size': 17})
plt.xticks([-4*pi, -2*pi, 0, 2*pi, 4*pi],['$-4\pi$','$-2\pi$','$0$','$2\pi$','$4\pi$'])
plt.yticks([0, 1],['$0$','$1$'])
plt.plot(x, np.sin(x)/x)
plt.ylim(-0.5, 1.5)
plt.show
```

## Add Title and Legend To the Graph

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import numpy as np
import matplotlib.pyplot as plt
pi = np.pi
x = np.linspace(-4*pi,4*pi,1000)
plt.rcParams['figure.figsize'] = (11,4)
plt.rcParams.update({'font.size': 17})
plt.xticks([-4*pi, -2*pi, 0, 2*pi, 4*pi],['$-4\pi$','$-2\pi$','$0$','$2\pi$','$4\pi$'])
plt.yticks([0, 1],['$0$','$1$'])
plt.plot(x, np.sin(x)/x, label = r'$f(x)=\frac{\sin(x)}{x}$')
plt.title('The $sinc$ Function')
plt.legend(loc = 'best',fontsize=20)
plt.ylim(-0.5, 1.5)
plt.show
```

## Add Axis to the Graph

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import numpy as np
import matplotlib.pyplot as plt
pi = np.pi
x = np.linspace(-4*pi,4*pi,1000)
plt.rcParams['figure.figsize'] = (11,4)
plt.rcParams.update({'font.size': 17})
plt.xticks([-4*pi, -2*pi, 0, 2*pi, 4*pi],['$-4\pi$','$-2\pi$','$0$','$2\pi$','$4\pi$'])
plt.yticks([0, 1],['$0$','$1$'])
plt.plot(x, np.sin(x)/x, label = r'$f(x)=\frac{\sin(x)}{x}$')
plt.title('The $sinc$ Function')
plt.legend(loc = 'best',fontsize=20)
plt.ylim(-0.5, 1.5)
plt.axhline(0,color='black',lw=1)
plt.show
```
